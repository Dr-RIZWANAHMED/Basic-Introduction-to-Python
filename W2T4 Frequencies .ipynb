### Libraries

```python
# LIBRARY
import numpy as np # vector manipulation

# THIS IS FOR PLOTTING
%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import warnings
warnings.filterwarnings("ignore")
```

## Sinewave
A sine wave describes a periodic signal of amplitude, $A$ and frequency $f$ Hz, over a time period $t$:
$$ A\sin(\pi \ f \  t). $$
The plot below shows a sinewave of amplitude 3 and a frequency of 2 Hz from 0 to 10 in steps of 0.01:
$$ 3\sin(\pi \ 2 \  t). $$


```python
time=np.arange(0,10,0.01)
Amplitude=3
freq=2

Sinewave=Amplitude*np.sin(np.pi*freq*time)


fig = plt.figure(figsize=(8,4)) # This setups the size of the figure
plt.plot(time,Sinewave,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.show() # This plots the figure
```

## Summation of Sinewaves
Summing sinewaves together results in interesting new signals. 
Here we add together three sinewaves over a period from 0 to 10 with 0.001 steps:
1. Amplitude 3 with a frequency of 0.5 Hz;
2. Amplitude 1 with a frequency of 2 Hz;
3. Amplitude 0.1 with a frequency of 9 Hz.
The plot below shows the first sinewave:
$$ 5 \sin(0.5 \pi t), $$
from 0 to 10.

```python
time=np.arange(0,10,0.001)


Amplitude1=3
freq1=0.5

Sinewave1=Amplitude1*np.sin(np.pi*freq1*time)

fig = plt.figure(figsize=(6,4)) # This setups the size of the figure

plt.plot(time,Sinewave1,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.title('Sinewave 1: Amplitude 3 with a frequency of 0.5 Hz ')
plt.show()
```

The plot below shows the second sinewave:
$$ 1 \sin(2 \pi t), $$
from 0 to 10.

```python
Amplitude2=1
freq2=2
Sinewave2=Amplitude2*np.sin(np.pi*freq2*time)

fig = plt.figure(figsize=(6,4)) # This setups the size of the figure

plt.plot(time,Sinewave2,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude ')
plt.title('Sinewave 2: Amplitude 1 with a frequency of 2 Hz')
plt.show()
```

The plot below shows the third sinewave:
$$ 0.1 \sin(9 \pi t), $$
from 0 to 10.

```python
Amplitude3=0.1
freq3=9
Sinewave3=Amplitude3*np.sin(np.pi*freq3*time)

fig = plt.figure(figsize=(6,4)) # This setups the size of the figure

plt.plot(time,Sinewave3,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude ')
plt.title('Sinewave 3: Amplitude 0.1 with a frequency of 9 Hz')
plt.show()
```

The plot below shows each individual sinewave and the summation of them in a grid.

```python
Sum_Sinewave=Sinewave1+Sinewave2+Sinewave3

fig = plt.figure(figsize=(12,8)) # This setups the size of the figure
plt.subplot(221)
plt.plot(time,Sinewave1,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.ylim([-4,4])
plt.title('Sinewave 1')


plt.subplot(222)
plt.plot(time,Sinewave2,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.ylim([-4,4])
plt.title('Sinewave 2')


plt.subplot(223)
plt.plot(time,Sinewave3,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.ylim([-4,4])
plt.title('Sinewave 3')

plt.subplot(224)
plt.plot(time,Sum_Sinewave,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.ylim([-4,4])
plt.title('Sum of Sinewaves')

plt.tight_layout()
plt.show() # This plots the figure
```
