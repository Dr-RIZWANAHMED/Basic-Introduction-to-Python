### Libraries

```python
# LIBRARY
import numpy as np # vector manipulation
# THIS IS FOR PLOTTING
%matplotlib inline
import matplotlib.pyplot as plt # side-stepping mpl backend
import warnings
warnings.filterwarnings("ignore")
```

## Summation of Sinewaves
The plot below shows a summation of Hsinewaves over a period from 0 to 10 with a sample rate sr=0.01 steps:
1. Amplitude 3 with a frequency of 0.5 Hz;
2. Amplitude 1 with a frequency of 2 Hz;
3. Amplitude 0.1 with a frequency of 9 Hz.
The summation is
$$3\sin(0.5\pi t)+1 \sin(2 \pi t)+0.1 \sin(9 \pi t) $$
is plotted below.

```python
sr=0.01
time=np.arange(0,10,sr)


Amplitude1=3
freq1=0.5
Sinewave1=Amplitude1*np.sin(np.pi*freq1*time)

Amplitude2=1
freq2=2
Sinewave2=Amplitude2*np.sin(np.pi*freq2*time)

Amplitude3=0.1
freq3=9
Sinewave3=Amplitude3*np.sin(np.pi*freq3*time)

Sum_Sinewave=Sinewave1+Sinewave2+Sinewave3

```

```python
fig=plt.figure(figsize=(6, 4))

plt.plot(time,Sum_Sinewave,'-',color='black')
plt.xlabel('time (s)')
plt.ylabel('Amplitude')
plt.ylim([-4,4])
plt.title('Sum of Sinewaves')

plt.tight_layout()
plt.show() # This plots the figure
```

## Fast Fourier Transform 
To decompse the signal into it's different frequency components we use the Fast Fourier Transform function __fft__.

```python
fft_x = np.fft.fft(Sum_Sinewave)
```

To convert the time axis to frequency we use the fftfreq function which takes the length of the signal n and the sample rate sr.

```python
n = len(fft_x)
freq = np.fft.fftfreq(n, sr)
```

```python
fig=plt.figure(figsize=(6, 4))
plt.plot(freq, 2.0/n *np.abs(fft_x))
plt.xlabel('Frequency')
plt.xlim([0.1,20])
plt.ylabel('PSD (dB)')
plt.grid()
plt.show()
```
